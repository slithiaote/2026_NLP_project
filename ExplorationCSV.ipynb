# Filtrer les documents disponibles et vérifier que la transcription existe

on a des infos sur le titulaire : profession, age
la question est comment faire des liaisons pour avoir un réseau : 
appartenance à un parti : sera très fragmenté / clusters
proximité des documents : nécessite de définir un threshold (faire de sorte à matcher un réseau de personnes, puis explorer le réseau pour voir si cela correspond aux caractéristiques d'un réseau de personnes, par exemple la transitivité)

```python
# select 1988 election (because we only have transcriptions for this election)
csv1988 = csv.loc[(csv['date'].dt.year==1988) & (csv['contexte-election']=='législatives')]
```

```python
# check that this corresponds to the legislatives
print(csv1988['contexte-election'].value_counts())
print(csv1988["contexte-tour"].value_counts())
```

```python
csv1988["id"]
```

```python
for row in csv1988.itertuples():
    with contextlib.suppress(FileNotFoundError):
        with open("text_files/1988/legislatives/"+row.id+".txt", "r") as file:
            content = file.read()
        csv1988.loc[row.Index,"transcript"] = content
        csv1988.loc[row.Index,"transcript_length"] = len(content)
```

```python
csv1988["transcript_length"].value_counts()
```

```python
with open("csv1988.pkl", "wb") as f:
    pickle.dump(csv1988, f)
```
