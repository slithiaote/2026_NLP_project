# Construction d'un nom normalisé / le nom figure-t-il dans le transcript ?

## Lowercase

```python
df = csv1988.select(pl.col("titulaire-nom"),pl.col("transcript"), 
                    pl.col("titulaire-nom").str.to_lowercase().alias("titulaire-lowercase"),
                    pl.col("transcript").str.to_lowercase().str.contains(pl.col("titulaire-nom").str.to_lowercase()).alias("titulaire-lowercase-is-in-transcript"))
```

```python
df.get_column("titulaire-lowercase-is-in-transcript").value_counts()
```

```python
df.filter(pl.col("titulaire-lowercase-is-in-transcript") != True).select(pl.col("titulaire-lowercase"),pl.col("transcript"))
```

On remarque que plus de 90% des transcripts contiennent bien le nom du titulaire. Les situations dans lesquelles le nom n'est pas retrouvé correspondent pour la plupart à des noms accentués.
